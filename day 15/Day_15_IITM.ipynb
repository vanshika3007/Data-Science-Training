```python
# ML:
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# Machine learning:
from sklearn import *
print('Done')
```

```python
data = {'Study_hrs':range(1,11),
       'Marks Obtained':[25,33,46,50,55,67,78,85,95,100]}

df = pd.DataFrame(data)
df
```

```python
# ML model build?: x, y: +ve, -ve
# Scatter Plot
plt.title('Study vs Marks')
plt.scatter(df['Study_hrs'],df['Marks Obtained'],marker = 'o', color = 'r')
plt.xlabel('Study Hrs')
plt.ylabel('Marks')
plt.show()
```

```python
# Training : Zero error
# Future: Predict:
```

```python
# # Y = mx + c
# # M : Slope
# # C : Intercept

# Y = Marks
# x = Study_hrs
```

```python
# x(1), y(1): simple Linear Regression
# x(multiple): y(1): Multiple Linear Regression

df
```

```python
# Import Module
from sklearn.linear_model import LinearRegression

# Object Model Creation
model = LinearRegression()


# X, y
X = df[['Study_hrs']]
y = df['Marks Obtained']


# Training/Learning
model.fit(X,y)
```

```python
m = model.coef_
c = model.intercept_
print(m,c)
```

```python

std_hrs = [[5.5]]

marks = m * std_hrs + c
# y = mx + c
# "y = β₀ + β₁x"
# y = wx +  c

print('Predicted Marks: ',marks[0][0])
```

```python
import warnings
warnings.filterwarnings('ignore')
study_hrs = input('Enter study hours: ')

std_hr = [[float(study_hrs)]]

ans = model.predict(std_hr)
marks = round(ans[0],2)

if marks > 100:
  marks = 100

elif marks<0:
  marks = 0

else:
  marks = marks

print(f'Your Predicted marks is: {marks}')
```

```python
# Description:
# The advertising dataset captures the sales revenue generated
# with respect to advertisement costs across multiple channels like radio, tv, and newspapers.
# It is required to understand the impact of ad budgets on the overall sales.
```

```python
# y = m*x + c

# X = 3 features

# y = m1*x1 + m2*x2 + m3*x3 + c

# y =  β₀+ β1*x1 + β2*x2 + β3*x3

# y = w1*x1 + w2*x2 + w3*x3 + c


# x1 = TV_budget
# x2 = radio
# x3 = newspaper


```

```python
X = df_multi.iloc[:,:-1]
y = df_multi['Sales']
```

```python
model = LinearRegression()

model.fit(X,y)
```
