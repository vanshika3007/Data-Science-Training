```python
class ATM:
    def __init__(self):
        print('object created successfully')
        self.pin = 1111
        self.balance = 2000

    def change_pin(self):
        self.pin = input('enter new PIN: ')
        print('You PIN has been changed!!')

    def show_balance(self):
        print(f'''Your Current Balance is Rs{self.balance} ''')

    def atm_performance_status(self):
        print('ATM working fine 100%')
```

```python
class ATM_VERSION_1:
    def __init__(self):
        print('object created successfully')
        self._pin = 1111#protected variable
        self._balance = 2000 #public variable

    def change_pin(self): #public method
        self._pin = input('enter new PIN: ')
        print('You PIN has been changed!!')

    def show_balance(self):
        print(f'''Your Current Balance is Rs{self._balance} ''')

    def atm_performance_status(self):
        print('ATM working fine 100%')
```

```python
axis = ATM_VERSION_1()
```

```python
axis._pin
```

```python
axis._balance
```

```python
class ATM_VERSION_2:
    def __init__(self):
        print('object created successfully')
        self.__pin = 1111#protected variable #single underscore for protected and double for private
        self.__balance = 2000 #public variable

    def change_pin(self): #public method
        self.__pin = input('enter new PIN: ')
        print('You PIN has been changed!!')

    def show_balance(self):
        print(f'''Your Current Balance is Rs{self.__balance} ''')

    def __atm_performance_status(self):#private method
        print('ATM working fine 100%')
```

```python
hdfc =  ATM_VERSION_2()
```

```python
hdfc.change_pin()
```

```python
hdfc._ATM_VERSION_2__pin
```

```python
hdfc._ATM_VERSION_2__balance
```

```python
hdfc._ATM_VERSION_2__atm_performance_status()
```
